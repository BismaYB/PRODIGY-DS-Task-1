<h3>To analyze population trends across all years, sum the total population in all countries for each year and
create a bar chart to visualize the trend. 
This approach gives a clear overview of global population growth over time.</h3>

```python
# Extract year columns (1960-2023)
year = df.columns[4:]  
total_population= df[year].sum().dropna()

# Plotting the global population trend over the years using bar chart
plt.figure(figsize=(14, 7))
plt.bar(total_population.index, total_population.values,color='black')
plt.title('Population Trends (1960-2023)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Population', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

<h3>Analysis of the Visualization</h3>
The bar chart above illustrates the global population trend from 1960 to 2023:

Trend:

A clear upward trajectory in the global population over the years.
Population growth appears steady with noticeable acceleration in later decades.

Key Insights:

Population was significantly smaller in the 1960s but has increased rapidly over the years.
The consistent growth highlights the global trend of population expansion.

```python
# Extract relevant columns: 'Country Name' and '2023'
population_2023 = df[['Country Name', '2023']].dropna()
population_2023 = population_2023.sort_values(by='2023', ascending=False).head(10)  # Top 10 most populated countries

# Plotting a bar chart for the top 10 most populated countries
plt.figure(figsize=(12, 6))
plt.bar(population_2023['Country Name'], population_2023['2023'], color='black')
plt.title('Top 10 Most Populated Countries in 2023', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Population ', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
